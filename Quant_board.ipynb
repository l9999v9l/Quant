### Definitions of Key Metrics

-   **Inflow**: Represents the total volume or value of a specific cryptocurrency moving into exchanges or a particular address/entity within a given timeframe. It can indicate buying pressure or deposits.
-   **Outflow**: Represents the total volume or value of a specific cryptocurrency moving out of exchanges or a particular address/entity within a given timeframe. It can indicate selling pressure or withdrawals.
-   **Exhaustion**: A metric indicating the depletion of buying or selling pressure, suggesting a potential reversal in price trends. For example, 'buying exhaustion' could mean most buyers have already entered the market, and there's little demand left to push prices higher.
-   **Sentiment**: The overall market mood or feeling towards a specific cryptocurrency or the market in general. It can be derived from social media mentions, news articles, or other qualitative data sources, often expressed as positive, negative, or neutral.
-   **Whale Inflow**: Similar to 'Inflow', but specifically tracks large transactions or movements of cryptocurrency by 'whales' (entities holding significant amounts of crypto). This can signal substantial buying interest or strategic moves by major players.
-   **Success Rate**: In the context of trading strategies or predictions, this could represent the percentage of trades that result in a profit, or the accuracy of a prediction model. For asset pairs, it might relate to the profitability of arbitrage opportunities or the effectiveness of a trading bot for that pair.

## Calculate Significant Velocity and Accuracy Hours

### Subtask:
Calculate the 'net flow' for each asset pair, then identify the timestamp with the highest absolute net flow (highest significant velocity) and the timestamp with the highest success rate (exact accuracy hour) for each asset.


**Reasoning**:
The subtask requires calculating 'net_flow', identifying the timestamp with the highest absolute net flow (significant velocity), and the timestamp with the highest success rate (accuracy hour) for each asset pair. This code block will perform these calculations and store the results.



````python
import pandas as pd

# 1. Calculate 'net_flow'
df_metrics['net_flow'] = df_metrics['inflow'] - df_metrics['outflow']

# Prepare lists to store results
highest_velocities = []
highest_success_rates = []

# 2. For each asset pair, identify the timestamp with the maximum absolute net_flow
for pair in df_metrics['asset_pair'].unique():
    df_pair = df_metrics[df_metrics['asset_pair'] == pair]

    # Highest significant velocity
    idx_max_velocity = df_pair['net_flow'].abs().idxmax()
    row_max_velocity = df_pair.loc[idx_max_velocity]
    highest_velocities.append({
        'asset_pair': pair,
        'highest_significant_velocity': row_max_velocity['net_flow'].abs(),
        'velocity_hour': row_max_velocity['timestamp']
    })

    # Highest success rate (accuracy hour)
    idx_max_success_rate = df_pair['success_rate'].idxmax()
    row_max_success_rate = df_pair.loc[idx_max_success_rate]
    highest_success_rates.append({
        'asset_pair': pair,
        'highest_success_rate': row_max_success_rate['success_rate'],
        'accuracy_hour': row_max_success_rate['timestamp']
    })

# Convert results to DataFrames for better readability
df_highest_velocities = pd.DataFrame(highest_velocities)
df_highest_success_rates = pd.DataFrame(highest_success_rates)

# 4. Display the identified 'highest significant velocities' and 'accuracy hours'
print("\n--- Highest Significant Velocities (Absolute Net Flow) ---")
print(df_highest_velocities)

print("\n--- Highest Success Rates (Accuracy Hours) ---")
print(df_highest_success_rates)
````

## Identify High Success Rate Wallets

### Subtask:
Analyze transaction data and apply the defined 'success rate' methodology to 'extract the highest significant success rate wallet(s)' from which insights can be drawn.


**Reasoning**:
The subtask requires simulating wallet performance data, so I'll create a new DataFrame `df_wallet_success` with unique wallet IDs, asset pairs, and simulated success rates. Then, I will identify the wallet with the highest success rate for each asset pair and store the results in `df_top_wallets`.



````python
import pandas as pd
import numpy as np

# Get unique asset pairs from the existing df_metrics DataFrame
asset_pairs = df_metrics['asset_pair'].unique()

# 1. Simulate unique wallet IDs
num_wallets = 50 # Let's simulate 50 wallets
wallet_ids = [f'wallet_{i:03d}' for i in range(num_wallets)]

# Create a list to store wallet performance data
wallet_data = []

# 2. Populate df_wallet_success by assigning random wallet_success_rate for each wallet-asset pair combination
for wallet in wallet_ids:
    for pair in asset_pairs:
        wallet_data.append({
            'wallet_id': wallet,
            'asset_pair': pair,
            'wallet_success_rate': np.random.uniform(0.2, 0.95) # Simulated success rate between 0.2 and 0.95
        })

df_wallet_success = pd.DataFrame(wallet_data)

# Create an empty list to store top wallets
top_wallets_list = []

# 3. For each unique asset_pair, identify the wallet_id with the highest wallet_success_rate
for pair in asset_pairs:
    df_pair_wallets = df_wallet_success[df_wallet_success['asset_pair'] == pair]
    # Get the row with the maximum wallet_success_rate for the current asset pair
    top_wallet_row = df_pair_wallets.loc[df_pair_wallets['wallet_success_rate'].idxmax()]

    top_wallets_list.append({
        'asset_pair': top_wallet_row['asset_pair'],
        'top_wallet_id': top_wallet_row['wallet_id'],
        'highest_wallet_success_rate': top_wallet_row['wallet_success_rate']
    })

# 4. Store these identified high success rate wallets in a new DataFrame
df_top_wallets = pd.DataFrame(top_wallets_list)

# 5. Print the df_top_wallets DataFrame to display the results
print("\n--- Top Wallets by Highest Success Rate per Asset Pair ---")
print(df_top_wallets.head())
print("\n--- df_wallet_success Info ---")
print(df_wallet_success.info())
````

## Organize and Prioritize Asset Pairs

### Subtask:
Group the calculated data by each asset pair and apply a prioritization logic to organize the 'highest priority pair up down' based on the analytical results.


**Reasoning**:
The subtask requires merging existing dataframes, normalizing specific columns, calculating a prioritization score, and then sorting the assets based on this score to identify the highest priority pairs. This code block will perform all these steps.



````python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# 1. Merge the DataFrames into a single DataFrame (df_prioritized_assets)
# Start with df_highest_velocities and merge df_highest_success_rates
df_prioritized_assets = pd.merge(
    df_highest_velocities,
    df_highest_success_rates[['asset_pair', 'highest_success_rate']],
    on='asset_pair',
    how='inner'
)

# Then merge df_top_wallets
df_prioritized_assets = pd.merge(
    df_prioritized_assets,
    df_top_wallets[['asset_pair', 'top_wallet_id', 'highest_wallet_success_rate']],
    on='asset_pair',
    how='inner'
)

# 2. Normalize the specified columns to a scale between 0 and 1
scaler = MinMaxScaler()

df_prioritized_assets['normalized_velocity'] = scaler.fit_transform(
    df_prioritized_assets[['highest_significant_velocity']]
)
df_prioritized_assets['normalized_accuracy_rate'] = scaler.fit_transform(
    df_prioritized_assets[['highest_success_rate']]
)
df_prioritized_assets['normalized_wallet_rate'] = scaler.fit_transform(
    df_prioritized_assets[['highest_wallet_success_rate']]
)

# 3. Create a new column named prioritization_score
df_prioritized_assets['prioritization_score'] = (
    df_prioritized_assets['normalized_velocity'] +
    df_prioritized_assets['normalized_accuracy_rate'] +
    df_prioritized_assets['normalized_wallet_rate']
)

# 4. Sort the df_prioritized_assets DataFrame in descending order
df_prioritized_assets = df_prioritized_assets.sort_values(
    by='prioritization_score', ascending=False
).reset_index(drop=True)

# 5. Display the df_prioritized_assets DataFrame
print("\n--- Prioritized Asset Pairs ---")
print(df_prioritized_assets)

# Display data types of the new dataframe
print("\n--- df_prioritized_assets Info ---")
print(df_prioritized_assets.info())
````

## Complete SQL Query

### Subtask:
Complete the provided incomplete SQL query by finishing the `hourly_metrics` CTE and adding a final `SELECT` statement that combines relevant data from all CTEs (hourly_whale_flows, wallet_performance, top_wallets, and hourly_metrics).

The subtask requires completing a provided SQL query. I will generate a code block containing the full SQL query as instructed.



````python
sql_query = """WITH hourly_whale_flows AS (
    SELECT
        DATE_TRUNC('hour', block_time) AS timestamp,
        token_symbol AS asset_pair,
        SUM(amount) AS whale_inflow
    FROM
        transactions_table
    WHERE
        sender_is_whale = TRUE
    GROUP BY
        1, 2
),
wallet_performance AS (
    SELECT
        wallet_address AS wallet_id,
        token_symbol AS asset_pair,
        AVG(CASE WHEN transaction_status = 'success' THEN 1.0 ELSE 0.0 END) AS wallet_success_rate
    FROM
        wallet_transactions_table
    GROUP BY
        1, 2
),
top_wallets AS (
    SELECT
        asset_pair,
        wallet_id AS top_wallet_id,
        wallet_success_rate AS highest_wallet_success_rate
    FROM (
        SELECT
            asset_pair,
            wallet_id,
            wallet_success_rate,
            ROW_NUMBER() OVER(PARTITION BY asset_pair ORDER BY wallet_success_rate DESC) as rn
        FROM
            wallet_performance
    )
    WHERE
        rn = 1
),
hourly_metrics AS (
    SELECT
        DATE_TRUNC('hour', block_time) AS timestamp,
        token_symbol AS asset_pair,
        SUM(CASE WHEN transaction_type = 'inflow' THEN amount ELSE 0 END) AS inflow,
        SUM(CASE WHEN transaction_type = 'outflow' THEN amount ELSE 0 END) AS outflow,
        AVG(market_exhaustion_score) AS exhaustion,
        AVG(sentiment_score) AS sentiment,
        AVG(trade_success_rate) AS success_rate
    FROM
        market_data_table
    GROUP BY
        1, 2
)
SELECT
    hm.timestamp,
    hm.asset_pair,
    hm.inflow,
    hm.outflow,
    hm.exhaustion,
    hm.sentiment,
    hm.success_rate,
    hwf.whale_inflow,
    tw.top_wallet_id,
    tw.highest_wallet_success_rate
FROM
    hourly_metrics hm
LEFT JOIN
    hourly_whale_flows hwf ON hm.timestamp = hwf.timestamp AND hm.asset_pair = hwf.asset_pair
LEFT JOIN
    top_wallets tw ON hm.asset_pair = tw.asset_pair;"""

print("SQL query completed and stored in 'sql_query' variable.")
````

## Recalculate Asset Prioritization

### Subtask:
Re-run the asset prioritization logic, specifically the steps that merge data and calculate the `prioritization_score`, as `df_top_wallets` has been updated with the simulated Dune data.


**Reasoning**:
The subtask requires recalculating the asset prioritization score using the updated `df_top_wallets`. This involves merging the relevant dataframes, normalizing specific columns, calculating the `prioritization_score`, and sorting the results as per the instructions.



````python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# 1. Merge the DataFrames into a single DataFrame (df_prioritized_assets)
# Start with df_highest_velocities and merge df_highest_success_rates
df_prioritized_assets = pd.merge(
    df_highest_velocities,
    df_highest_success_rates[['asset_pair', 'highest_success_rate']],
    on='asset_pair',
    how='inner'
)

# 2. Then merge df_top_wallets (which has been updated with Dune data)
df_prioritized_assets = pd.merge(
    df_prioritized_assets,
    df_top_wallets[['asset_pair', 'top_wallet_id', 'highest_wallet_success_rate']],
    on='asset_pair',
    how='inner'
)

# 3. Normalize the specified columns to a scale between 0 and 1
scaler = MinMaxScaler()

df_prioritized_assets['normalized_velocity'] = scaler.fit_transform(
    df_prioritized_assets[['highest_significant_velocity']]
)
df_prioritized_assets['normalized_accuracy_rate'] = scaler.fit_transform(
    df_prioritized_assets[['highest_success_rate']]
)
df_prioritized_assets['normalized_wallet_rate'] = scaler.fit_transform(
    df_prioritized_assets[['highest_wallet_success_rate']]
)

# 4. Create a new column named prioritization_score
df_prioritized_assets['prioritization_score'] = (
    df_prioritized_assets['normalized_velocity'] +
    df_prioritized_assets['normalized_accuracy_rate'] +
    df_prioritized_assets['normalized_wallet_rate']
)

# 5. Sort the df_prioritized_assets DataFrame in descending order
df_prioritized_assets = df_prioritized_assets.sort_values(
    by='prioritization_score', ascending=False
).reset_index(drop=True)

# 6. Display the head and info of the updated df_prioritized_assets DataFrame
print("\n--- Recalculated Prioritized Asset Pairs (Head) ---")
print(df_prioritized_assets.head())
print("\n--- df_prioritized_assets Info ---")
print(df_prioritized_assets.info())
````
